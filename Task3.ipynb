```python
#!/usr/bin/env python3

import re
import numpy as np
import pandas as pd
import openai
# Web text scraper
from urllib.request import urlopen
from bs4 import BeautifulSoup

def data_reader(dir_link):
    data = pd.read_csv(dir_link, sep=';', quotechar='"', encoding='utf-8', na_values=' ')
    return data

def web_text_extractor(url):
    try:
        html = urlopen(url).read()
        soup = BeautifulSoup(html, features="html.parser")
        # Extract the text within the <div> element
        content_div = soup.find('div', id='LangereOmschrijving')
        text = content_div.get_text()
        text = text.strip()
        text = re.sub(r'\n', ' ', text)
        text = re.sub(r'\n\s*\n', '\n', text)
        return text
    except:
        return True


def check_porsche(text):
    pattern = r'Porsche'
    match = re.search(pattern, text, re.IGNORECASE)

    if match:
        return True
    else:
        return False

parruzi_url = 'https://www.paruzzi.com/uk/'

# Read data
df = data_reader('paruzzi_processed_v19.csv')

# Create Column
df['Paruzzi Desc.'] = np.nan
df['Paruzzi Web Link'] = np.nan
df['Paruzzi Desc.'] = df['Paruzzi Desc.'].astype(str)

for row_index, row_data in df.iterrows():
    print(f'Completed {row_index} of {len(df.index)}')

    cell_val = df['Paruzzi nr'].values[row_index]
    txt = web_text_extractor(parruzi_url+str(cell_val))
    if txt == True:
        df['Paruzzi Web Link'].values[row_index] = 0
        df.at[row_index, 'Porsche'] = 0
        df['Paruzzi Desc.'].values[row_index] = 0
        continue
    df['Paruzzi Web Link'].values[row_index] = 1
    df['Paruzzi Desc.'].values[row_index] = txt

    if check_porsche(txt):
        df.at[row_index, 'Porsche'] = 1
    else:
        df.at[row_index, 'Porsche'] = 0

# Save files
df.to_csv('paruzzi_processed_v20.csv',sep=';',index=False)

display(df)
```
